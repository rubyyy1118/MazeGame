```python
## the following line is just MAGIC ##############################################
## we need it to be able to draw images on the screen ############################
from medialib import *
initialize() # always the first instruction of the program
##################################################################################

##  each item of the list representing the background of each row
background=[['X','X','X','X','X','X','X','X','X','X','X','X','X','X','X','X','X','X'],
            ['X','_','_','L','_','_','M','_','_','E','_','_','J','_','_','_','_','X'],
            ['X','_','_','_','M','_','B','_','L','_','_','_','_','B','_','J','_','X'],
            ['X','_','_','B','_','_','_','_','_','J','_','_','E','_','L','_','_','X'],
            ['X','X','X','X','X','X','X','X','X','X','X','X','X','X','X','X','X','X']] 

collidables = 'X' ## the letters that the avatar cannot pass
monsters='M'## the letters that the avatar will dead when meeting it

clear(200,200,200)## background


#draw the background
width=64
pic_size=30
def draw_background(background, player_x, player_y ,width):
    for row in background:
        for i in row:
            background_pic = i
            if background_pic == "X":
                draw('X.png', player_x, player_y)
            elif background_pic == "L":
                draw('L.png', player_x, player_y)
            elif background_pic == "M":
                draw('M.png', player_x,player_y)
            elif background_pic == "E":
                draw('E.png', player_x,player_y)
            elif background_pic == "J":
                draw('J.png', player_x,player_y)
            elif background_pic == "B":
                draw('B.png', player_x,player_y)
            elif background_pic == "_":
                draw('_.png', player_x,player_y)
            player_x += pic_size
        player_y += pic_size
        player_x = 10


## the first image is at (10,200)
x0=10
y0=200 

#Redraw the background to cover the paths already traversed by the character
def re_draw_background():
    clear(200,200,200)
    x0=10
    y0=200
    draw_background(background, x0,y0,width)
    text('Maze game',300,100,20)
    text('You are a lost hero, and have to find your way back home...',10,500,16)
    text('When you collect 20 points, you can get back to home!',10,520,16)
    text('To play: press a and d keys to move left and right',10,540,16)
    text('         press w and s keys to move upwards and downwards',10,560,16)
    text('         press q to quit the game',10,580,16)
 
point = 0
received_point = False

point = 0
airline={"J":"Jet2", "E":"EasyJet", "L":"Lufthansa"}


def player_move(man_x, man_y):
    global point
    draw("H.png",man_x , man_y)

    if background[(man_y - y0) // pic_size][(man_x - x0) // pic_size] in ["J", "E", "L"]:
        airline_key = background[(man_y - y0)//pic_size][(man_x - x0)//pic_size]
        text(f"Welcome on board with {airline[airline_key]}! You just get 1 point!", 100, 150, 20)
        point+=1
        background[(man_y - y0) // pic_size][(man_x - x0) // pic_size]= "_"

    elif background[(man_y - y0) // pic_size][(man_x - x0) // pic_size] == 'M':
        clear(200,200,200)
        text("Game over!",150,300,70)
        text("Current Points:{}".format(point),350,400,30)
        wait_mouse_leftclick()
        all_done()

    elif background[(man_y - y0) // pic_size][(man_x - x0) // pic_size] == 'B':
        background[(man_y - y0) // pic_size][(man_x - x0) // pic_size] = "_"
        point += 1
        clear(200, 200, 200)
        point_text = f"Current Points: {point}"
        text(point_text, 10, 10, 25)
        text("Welcome on board with British Airways! You just received 1 point!", 10, 50, 16)
        draw('MazeGame_imgs/I.png', 10, 200)
        text("However, now British Airways has an employee work crisis.", 10, 480, 20)
        text("If you help them to make the decision, you can get extra points!", 10, 500, 20)
        text("Please make your decision by inputting 'h' or 'i'", 10, 560, 20)
        text("    h. Hire new staff", 10, 520, 20)
        text("    i. Increase existing staff pay", 10, 540, 20)
        while True:
            choice = wait_key_press()
            if choice == "h":
                re_draw_background()
                text('Thank you for supporting hiring new staffs! You received extra 3 points!', 25, 150, 18)
                point += 3
                break
            elif choice == 'i':
                re_draw_background()
                text('Thank you for supporting increasing existing staff pay! You received extra 5 points!', 25, 150, 18)
                point += 5
                break
        
        draw("H.png", man_x, man_y)  # Ensure the character's position is updated after redrawing the background

man_x = x0+pic_size
man_y = y0+pic_size
game_running = True

def close_game():
    clear(200, 200, 200) 
    text("Game over!", 150, 300, 70)
    text("Current Points:{}".format(point), 10, 10, 25)
    wait_mouse_leftclick()  
    all_done()  


while game_running:
    re_draw_background()
    draw("H.png", man_x, man_y)

    while True:
        key = wait_key_press()

        if key == "a":
            if man_x < 70:
                continue
            else:
                re_draw_background()
                man_x -= pic_size
                player_move(man_x, man_y)
                text("Current Points:{}".format(point),10,10,25)
        
        elif key == "d":
            if man_x > 460:
                continue
            else:
                re_draw_background()
                man_x += pic_size
                player_move(man_x, man_y)
                text("Current Points:{}".format(point),10,10,25)
        
        elif key == "w":
            if man_y < 240:
                continue
            else:
                re_draw_background()
                man_y -= pic_size
                player_move(man_x, man_y)
                text("Current Points:{}".format(point),10,10,25)

        elif key == "s":
            if man_y > 270:
                continue
            else:
                re_draw_background()
                man_y += pic_size
                player_move(man_x, man_y)
                text("Current Points:{}".format(point),10,10,25)

        elif key == "q":
            close_game()  # Call the 'close_game' function to perform cleanup and close when ending the game
            game_running = False
            break
        else:
            text("Please give a valid input", 20,400,20)
            continue

        if point >= 20:
            clear(200,200,200)
            draw("you_win.png",100,50)
            text("Final Points:{}".format(point),300,500,30)
            wait_mouse_leftclick()
            all_done()
            game_running = False  # Set game_running to False to exit the loop
            break  # Exit the inner loop


# Ensure Pygame cleanup outside the loops
all_done()
```
